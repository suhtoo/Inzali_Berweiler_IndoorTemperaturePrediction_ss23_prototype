```python
# Load the trained ML model
filename = "/Users/hiz/Desktop/RWTH_3rdSem/CR_Protyping Project DSS - BIM/Project/Inzali_Berweiler_IndoorTemperaturePrediction_ss23_prototype/3_modelTrainingModule/rf_model.joblib"
rf_model = joblib.load(filename)
```

```python
# Assuming you have a function to preprocess the data and get roomTemp predictions
def get_roomTemp_predictions(df):
    features = df[['windspeed', 'sunHeight', 'weatherTemp', 'windowStatus']]
    roomTemp_predictions = rf_model.predict(features)
    return roomTemp_predictions

# Get the roomTemp predictions using the generated data frame
roomTemp_predictions = get_roomTemp_predictions(df)

# Add the predictions to the data frame
df['Predicted roomTemp'] = roomTemp_predictions

df.sample(10)
```
