```python
# Set the random seed for reproducibility
np.random.seed(42)

# Function to generate random data for the features
def generate_random_data(start_time, end_time):
    time_index = pd.date_range(start=start_time, end=end_time, freq='10S')
    num_samples = len(time_index)

    windspeed = np.random.uniform(0, 10, num_samples)
    sunHeight = np.random.uniform(0, 90, num_samples)
    weatherTemp = np.random.uniform(250, 320, num_samples)
    windowStatus = np.random.choice([1, 2, 3], num_samples)

    data = {
        'windspeed': windspeed,
        'sunHeight': sunHeight,
        'weatherTemp': weatherTemp,
        'windowStatus': windowStatus
    }

    df = pd.DataFrame(data, index=time_index)
    return df

# Generate a random start time as the current datetime
current_time = datetime.now()
start_time = current_time.replace(second=0, microsecond=0)
end_time = start_time + timedelta(minutes=10)

# Generate the random data frame
df = generate_random_data(start_time, end_time)

# Save the random data frame to a CSV file
df.to_csv("random_data.csv")
df.sample(10)
```
